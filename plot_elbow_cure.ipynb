```python
def obtain_errors(season,num_representatives):
    if season !=4:
        errors = spark.read.csv('./CURE_Errors/season='+str(season)+'_num_reps='+str(num_representatives),header=True,inferSchema=True)
    else:
        errors = spark.read.csv('./CURE_Errors/season=EVERYTHING_num_reps='+str(num_representatives),header=True,inferSchema=True)
    fin = errors.orderBy('k').select('error').toPandas().to_numpy()
    print(fin)
    return fin
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
#figure(figsize=(200,200), dpi=60)
# elbow plot
# we have num_reps 4,5,6,7
season = 0
#fig, axs = plt.subplots(4)
#fig.suptitle('CURE: Season=0 (Winter)')
for i in range(4,8):
    plt.plot([j for j in range(3,10)],obtain_errors(season,i))
    plt.xlabel("Number of Clusters")
    plt.ylabel("Sum of Squared Error")
    plt.title("CURE: Season=0 (Winter) Num Reps: " + str(i))
    plt.show()


#plt.plot()

```
